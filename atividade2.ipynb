```python
import matplotlib.pyplot as plt

# Definindo os horários (eixo X) e as temperaturas reais (eixo Y)
horarios = list(range(25))  # De 0 a 24
temperaturas = [22, 21, 20, 20, 19, 19, 20, 21, 22, 23, 24, 25, 26, 27, 27, 26, 25, 24, 23, 22, 21, 20, 20, 19, 18]

# Criando o gráfico de linha
plt.plot(horarios, temperaturas, marker='o')

# Personalizando o gráfico
plt.title('Evolução da Temperatura Durante o Dia em Curitiba, PR')
plt.xlabel('Horários (h)')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.annotate('Temperatura Máxima ao Meio-Dia', xy=(12, 27), xytext=(15, 28),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Exibindo o gráfico
plt.show()
```
