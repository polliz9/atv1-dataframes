```python
# Importando bibliotecas necessárias
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Gerando dados fictícios
np.random.seed(42)
num_vehicles = 200

# Potência do motor (em cavalos) e peso do veículo (em kg)
engine_power = np.random.uniform(50, 300, num_vehicles)
vehicle_weight = np.random.uniform(800, 2500, num_vehicles)

# Classificação do veículo (1 = Econômico, 0 = Não Econômico)
# Assumindo que veículos com potência menor que 150 cv e peso menor que 1500 kg são econômicos
labels = ((engine_power < 150) & (vehicle_weight < 1500)).astype(int)

# Criando um DataFrame
data = pd.DataFrame({'Engine Power': engine_power, 'Vehicle Weight': vehicle_weight, 'Economic': labels})

# Dividindo os dados em conjuntos de treino e teste
X = data[['Engine Power', 'Vehicle Weight']]
y = data['Economic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando e treinando o modelo de árvore de decisão
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Visualizando as regras geradas pela árvore
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=['Engine Power', 'Vehicle Weight'], class_names=['Não Econômico', 'Econômico'], filled=True)
plt.show()
```
