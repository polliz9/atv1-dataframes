```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictcios de desempenho de vendas
dados = {
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta', 'Sbado', 'Domingo'],
    'Vendas': [1200, 1500, 1800, 2200, 3000, 4000, 3200],
    'Clientes': [300, 400, 500, 550, 700, 850, 600],
    'Lucro': [300, 375, 450, 550, 750, 1000, 800]
}

df = pd.DataFrame(dados)

# Grfico de Barras: Total de Vendas por Dia
plt.figure(figsize=(10, 6))
sns.barplot(x='Dia', y='Vendas', data=df)
plt.title('Total de Vendas por Dia')
plt.xlabel('\nDia da Semana\n')
plt.ylabel('Vendas (R$)')
plt.show()

# Anlise do Grfico de Barras:
# O grfico de barras mostra que as vendas aumentam gradualmente ao longo da semana,
# atingindo o pico no sbado e diminuindo ligeiramente no domingo.
# Isso sugere que os dias de fim de semana so os mais movimentados em termos de vendas.

# Grfico de Disperso: Nmero de Clientes vs Total de Vendas
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Clientes', y='Vendas', data=df)
plt.title('Nmero de Clientes vs Total de Vendas')
plt.xlabel('Nmero de Clientes\n\n')
plt.ylabel('Vendas (R$)')
plt.show()

# Anlise do Grfico de Disperso:
# O grfico de disperso demonstra uma correlao positiva entre o nmero de clientes e o total de vendas.
# Isso indica que,  medida que o nmero de clientes aumenta, as vendas tambm tendem a aumentar.
# A relao parece ser bastante linear, sugerindo uma forte dependncia entre as duas variveis.

# Heatmap: Correlao entre Vendas, Clientes e Lucro
plt.figure(figsize=(10, 6))
# Select only numerical columns for correlation calculation
correlacao = df[['Vendas', 'Clientes', 'Lucro']].corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlao entre Vendas, Clientes e Lucro')
plt.show()

# Anlise do Heatmap:
# O heatmap confirma a forte correlao positiva entre "Vendas" e "Clientes",
# assim como entre "Vendas" e "Lucro", e entre "Clientes" e "Lucro".
# Os valores prximos a 1 indicam uma correlao positiva muito forte,
# o que significa que essas variveis tendem a aumentar ou diminuir juntas.
```
