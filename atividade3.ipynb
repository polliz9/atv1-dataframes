```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictícios de desempenho de vendas
dados = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Vendas': [1200, 1500, 1800, 2200, 3000, 4000, 3200],
    'Clientes': [300, 400, 500, 550, 700, 850, 600],
    'Lucro': [300, 375, 450, 550, 750, 1000, 800]
}

df = pd.DataFrame(dados)

# Gráfico de Barras: Total de Vendas por Dia
plt.figure(figsize=(10, 6))
sns.barplot(x='Dia', y='Vendas', data=df)
plt.title('Total de Vendas por Dia')
plt.xlabel('\nDia da Semana\n')
plt.ylabel('Vendas (R$)')
plt.show()

# Análise do Gráfico de Barras:
# O gráfico de barras mostra que as vendas aumentam gradualmente ao longo da semana,
# atingindo o pico no sábado e diminuindo ligeiramente no domingo.
# Isso sugere que os dias de fim de semana são os mais movimentados em termos de vendas.

# Gráfico de Dispersão: Número de Clientes vs Total de Vendas
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Clientes', y='Vendas', data=df)
plt.title('Número de Clientes vs Total de Vendas')
plt.xlabel('Número de Clientes\n\n')
plt.ylabel('Vendas (R$)')
plt.show()

# Análise do Gráfico de Dispersão:
# O gráfico de dispersão demonstra uma correlação positiva entre o número de clientes e o total de vendas.
# Isso indica que, à medida que o número de clientes aumenta, as vendas também tendem a aumentar.
# A relação parece ser bastante linear, sugerindo uma forte dependência entre as duas variáveis.

# Heatmap: Correlação entre Vendas, Clientes e Lucro
plt.figure(figsize=(10, 6))
# Select only numerical columns for correlation calculation
correlacao = df[['Vendas', 'Clientes', 'Lucro']].corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlação entre Vendas, Clientes e Lucro')
plt.show()

# Análise do Heatmap:
# O heatmap confirma a forte correlação positiva entre "Vendas" e "Clientes",
# assim como entre "Vendas" e "Lucro", e entre "Clientes" e "Lucro".
# Os valores próximos a 1 indicam uma correlação positiva muito forte,
# o que significa que essas variáveis tendem a aumentar ou diminuir juntas.
```
